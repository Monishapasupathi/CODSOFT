```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:/Users/monis/Downloads/archive.csv')
```

```python
data.head(5)
```

```python
data.tail(5)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
x=data[['sepal_length','sepal_width','petal_length','petal_width']]
x
```

```python
y=data['species']
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4,random_state=30)
```

```python
std_scale=StandardScaler()
```

```python
std_scale
```

```python
xtrain=std_scale.fit_transform(xtrain)
xtest=std_scale.fit_transform(xtest)
```

```python
xtrain
```

```python
xtest
```

```python
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
y_pred=knn.predict(xtest)
```

```python
y_pred
```

```python
acc=accuracy_score(ytest,y_pred)
print("Accuracy:",acc)
```

```python
print("Classification Report")
```

```python
print(classification_report(ytest,y_pred))
```

```python
y_pred=knn.predict(xtest)
confu_matrix=confusion_matrix(ytest,y_pred)
```

```python
print("confusion matrix")
print(confu_matrix)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(confu_matrix,cmap='Blues')
plt.xlabel('predicted labels')
plt.ylabel('actual labels')
plt.title('confusion matrix')
plt.show()
```
